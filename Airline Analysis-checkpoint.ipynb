## Univariate Analysis

1. What do coach ticket prices look like? What are the high and low values? What would be considered the average? Does $500 seem like a good price for a coach ticket?

````python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels
import matplotlib.pyplot as plt
import math
import scipy.stats as stats
from scipy.stats import f_oneway, pearsonr, ttest_1samp

## Read in Data
flight = pd.read_csv("flight.csv")
print(flight.head())

## Task 1
print(len(flight))
print(flight.coach_price.max(), flight.coach_price.min())
sns.histplot(flight.coach_price)
plt.title('coach price distribution')
plt.figtext(.25, -.05, 'Appendix Figure 1: coach price distribution', fontsize=11)
print('the mean of the coach prices is ${}'.format(round(flight.coach_price.mean(),2)))
plt.savefig('D:/Data/airline_analysis/charts/1_price_vs_count.PNG',bbox_inches='tight')
````

- the data is generally normal
- prices range from 593 to 44

You want to better understand how often there are large delays so you can correctly set up connecting flights. What kinds of delays are typical?

## Bivariate Analysis

4. Create a visualization that shows the relationship between coach and first-class prices. What is the relationship between these two prices? Do flights with higher coach prices always have higher first-class prices as well?

````python
## Task 4
percent = .1
flight_sub = flight.sample(n=int(flight.shape[0]*percent))
sns.lmplot(data=flight_sub, x='coach_price', y='firstclass_price', line_kws={'color': 'black'})
plt.figtext(.25, -.05, 'Appendix Figure 5: coach price vs first class price', fontsize=11)
plt.title('coach price vs first class price')
plt.savefig('D:/Data/airline_analysis/charts/4_coach_price_vs_first_class_price.PNG',bbox_inches='tight')
plt.show()

print(np.cov(flight.firstclass_price, flight.coach_price))
corr_pass_price, p = pearsonr(flight.firstclass_price, flight.coach_price)
print(corr_pass_price)
````

- there is a linear relationship between coach price and first class price

### Linear Regression

````python
import statsmodels.api as sm
plt.scatter(flight.passengers, flight.coach_price)
model = sm.OLS.from_formula('coach_price ~ passengers', flight)
results = model.fit()
print(results.params)
````

````python
fitted_values = results.predict(flight)
residuals = flight.coach_price - fitted_values
plt.hist(residuals)
````

````python
plt.scatter(fitted_values, residuals)
plt.axhline(y=0, color= 'black', linestyle = '--')
````

- homoscedasticity is possibly met
